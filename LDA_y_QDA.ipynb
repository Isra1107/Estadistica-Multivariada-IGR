```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
mu1 = [-4, 4]
mu2 =[3,-3]
sigma1 = np.array([[10, 1], [1, 5]]) ##Estas matrices son sim√©tricas y positivas definidas
sigma2 = np.array([[3, 0], [0, 4]])

X = np.random.multivariate_normal(mu1, sigma1, 5000)
y = np.ones((5000,))
y[:2000] = 0
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 10))
plt.scatter(X[:, 0], X[:, 1], s=3, c=list(map(lambda x:'blue' if x==0 else 'red', y)))
plt.show()
```

```python
def f_x_y_1(x):
    const = 1 / (2 * np.pi * np.power(np.linalg.det(sigma1), 1/2))
    return const * np.exp(-0.5 * (x - mu1).T @ np.linalg.inv(sigma1) @ (x - mu1))[0, 0]

def f_x_y_2(x):
    const = 1 / (2 * np.pi * np.power(np.linalg.det(sigma2), 1/2))
    return const * np.exp(-0.5 * (x - mu2).T @ np.linalg.inv(sigma2) @ (x - mu2))[0, 0]
```

```python
lda = LinearDiscriminantAnalysis(n_components=1, solver='svd', store_covariance=True).fit(X, y)
a=lda.coef_
a[0,1]
```
